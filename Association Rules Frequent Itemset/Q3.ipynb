```python
# Import Package
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# TransactionEncoder is not necessary, since pd.get_dummies returns a dataframe with value 1 / 0
# from mlxtend.preprocessing import TransactionEncoder

## 讓DataFrame 能呈現的寬度大一點
pd.options.display.max_colwidth = 200

data = pd.read_csv("./hw1_data/customer.csv")
features = [
    'state_province',
    'yearly_income',
    'gender',
    'total_children',
    'num_children_at_home',
    'education',
    'occupation',
    'houseowner',
    'num_cars_owned'
]

df = pd.get_dummies(data[features], columns=features)
```

```python
apriori_items = apriori(df, min_support=0.05, use_colnames = True)
apriori_items
```

```python
rules = association_rules(apriori_items, metric = 'confidence', min_threshold=0.9)
rules = rules.sort_values(['confidence', 'support'],ascending=False)
rules.head(10)
```
