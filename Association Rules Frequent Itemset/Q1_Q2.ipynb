```python
# load dataset 1-11
df_dataset_other= pd.read_csv("sales_fact_1998.csv",)
# load dataset 12
df_dataset_dec= pd.read_csv("sales_fact_dec_1998.csv",)
# load dataset Product 
df_dataset_product= pd.read_csv("product.csv",)
```

```python
# concat sales_fact_1998 & sales_fact_dec_1998
df_dataset = pd.concat([df_dataset_other,df_dataset_dec],axis=0)
print(df_dataset)
```

```python
# join Product 
df_dataset = df_dataset.join(df_dataset_product.set_index('product_id'), on='product_id')
```

```python
# transaction_dataset
transaction_dataset = df_dataset.groupby(['customer_id','time_id'])
transaction_dataset.groups
```

```python
transaction_list = []

for name, group in transaction_dataset:
  transaction_products = group['product_name']
  transaction_list.append(transaction_products.tolist()) 
```

```python
transactionEncoder = TransactionEncoder()
tranEncoder_data = transactionEncoder.fit(transaction_list).transform(transaction_list)
transaction_df = pd.DataFrame(tranEncoder_data,columns=transactionEncoder.columns_)
transaction_df = transaction_df.replace(False,0)
transaction_df
```

1. 請利用 Apriori 演算法，從 Foodmart 資料庫的交易資料中，探勘符合 Minimum Support = 0.0001 且 Minimum Confidence = 0.9 的 Association Rules，並列出 Confidence 最高的前 10 條 Rules。若無法跑出結果，請簡述其原因。

```python
# apriori support >= 0.0001  
apriori_items = apriori(transaction_df, min_support=0.0001, use_colnames = True)
apriori_items
```

```python
# Association Rule confidence >= 0.9
Association_Rules = association_rules(apriori_items, metric = 'confidence', min_threshold=0.9)
Association_Rules = Association_Rules.sort_values(['support','confidence'],ascending=False)
Association_Rules_topten = Association_Rules.head(10)
Association_Rules_topten.to_csv("apriori_items.csv")
```

2. 請利用 FP-Growth 演算法，從 Foodmart 資料庫的交易資料中，探勘符合 Minimum Support = 0.0001 且 Minimum Confidence = 0.9 的 Association Rules，並列出 Confidence 最高的前 10 條 Rules。若無法跑出結果，請簡述其原因。



```python
# fpgrowth support >= 0.0001  
fpgrowth_items = fpgrowth(transaction_df, min_support=0.0001, use_colnames=True)
fpgrowth_items
```

```python
# Association Rule confidence >= 0.9
Association_Rules = association_rules(fpgrowth_items, metric = 'confidence', min_threshold=0.9)
Association_Rules = Association_Rules.sort_values(['support','confidence'],ascending=False)
Association_Rules_topten = Association_Rules.head(10)
Association_Rules_topten.to_csv("fpgrowth_items.csv")
```
